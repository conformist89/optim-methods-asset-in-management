```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cvxpy as cp
from scipy.optimize import minimize 
```

```python
daily_returns = pd.read_excel("../FERM_3_Assignment_2.xlsx", sheet_name="Daily_Returns_Raw_Data")
```

```python
daily_returns.head()
```

```python
daily_returns_only = daily_returns.drop(columns=["date"])
daily_returns_only.head()
```

```python
daily_losses = - daily_returns_only
daily_losses.head()
```

## We can see that ACB has the largest variance 

```python
sorted_losses = daily_losses.var().sort_values(ascending=False)
sorted_losses
```

```python
len(daily_losses)
```

```python
p_90 = 0.9
```

```python
def var_like_numpy(s, p=p_90):
    s = s.dropna().sort_values()
    N = len(s)
    Kp = int(np.ceil(p * N))
    return s.iloc[Kp - 1]
```

## Now we can also see that ACB has the largest VaR

```python
VaR = daily_losses.apply(var_like_numpy)
VaR.sort_values(ascending=False)
```

```python
def get_cvar(s, p=p_90):
    s = s.dropna().sort_values()
    N = len(s)

    coef = ((1 - p_90)*N)
    
    Kp = int(np.ceil(p * N))          
    tail = s.iloc[Kp - 1:]            # losses at/above VaR (worst tail)

    return tail.mean()
```

## We can also see that ACB has the largest CVaR

```python
CVaR = daily_losses.apply(get_cvar)
CVaR.sort_values(ascending=False)
```

## $VaR_{90}$ of the portfolio

```python
daily_returns_port = pd.read_excel("../FERM_3_Assignment_3.xlsx", sheet_name="Daily_Returns_Raw_Data")
```

```python
daily_returns_port.head()
```

```python
daily_returns_only_port = daily_returns_port.drop(columns=["date"])
daily_returns_only_port.head()
```

```python
mean_daily_ret_port = daily_returns_only_port.mean(axis=1)
mean_daily_ret_port
```

```python
mean_daily_loss_port = mean_daily_ret_port
mean_daily_loss_port
```

```python
def get_numpy_var(arr, p):
    N = len(arr)
    Kp = int(np.ceil((1-p) * N))
    
    sorted_losses = np.sort(arr)
    VaR = sorted_losses[Kp - 1]  
    return VaR
```

```python
p_90 = 0.9
```

```python
round(get_numpy_var(mean_daily_loss_port, p_90), 4)
```

```python
plt.hist(mean_daily_loss_port, bins = 100 )
plt.xlabel("Portfolio P&L")
plt.axvline(x=get_numpy_var(mean_daily_loss_port, p_90), color='g', label=r'$VaR_{90}$')
plt.legend()
```

## $CVaR_{90}$ of the portfolio

```python
def return_cvar(pnl, p):
    N = len(pnl)
    pnl_sorted = np.sort(pnl)

    n_tail = max(1, int(np.ceil((1 - p) * len(pnl_sorted))))
    return pnl_sorted[:n_tail].mean()
```

```python
round(return_cvar(mean_daily_loss_port, p_90), 4)
```

```python
plt.hist(mean_daily_loss_port, bins = 100 )
plt.xlabel("Portfolio P&L")
plt.axvline(x=get_numpy_var(mean_daily_loss_port, p_90), color='g', label=r'$VaR_{90}$')
plt.axvline(x=return_cvar(mean_daily_loss_port, p_90), color='r', label=r'$CVaR_{90}$')
plt.legend()
```
